# Tests Concerning Correlation

Các giả thuyết thường gặp nhất liên quan đến hệ số tương quan chính là so sánh hệ số tương quan giữa các tổng thể với 0, bởi vì chúng ta thường xuyên phải trả lời câu hỏi rằng có mối liên hệ nào tồn tại hay không - ngụ ý rằng giả thuyết không sẽ là hệ số tương quan bằng 0

Các giả thuyết không của kiểm định hệ số tương quan có thể là: (1) $\rho = 0$; (2) $\rho \le 0$; (3) $\rho \ge 0$

## Parametric Test

Hệ số tương quan theo cặp (`pairwise correlation coefficient`) thường được gọi là tương quan Pearson, tương quan hai biến (`bivariate correlation`) hoặc đơn giản là hệ số tương quan. Xem xét hai biến $X, Y$, hệ số tương quan mẫu của hai biến này là:

$$
r_{XY} = \dfrac{s_{XY}}{s_X s_Y}
$$

Nếu cả hai biến $X,Y$ đều tuân theo phân phối chuẩn, chúng ta có thể kiểm định các giả thuyết không ở trên thông qua hệ số tương quan mẫu $r$. Giá trị thống kê $t$ được tính như sau:

$$
t_{n-2} = \dfrac{r\sqrt{n-2}}{\sqrt{1-r^2}}
$$

## Noparametric Test

Nếu phân tích viên tin rằng các tổng thể đang được xem xét không tuân theo phân phối chuẩn, anh ta có thể sử dụng kiểm định dựa trên hệ số tương quan xếp hạng Spearman, $r_S$. Hệ số này về cơ bản tương đồng với hệ số tương quan Pearson, nhưng nó được tính toán dựa trên thứ hạng của hai biến trong các mẫu tương ứng. Phân tích viên xác định $r_S$ thông qua quy trình gồm 3 bước:

1. Đánh dấu quan sát của hai biến $X,Y$ từ lớn nhất đến nhỏ nhất. Gán giá trị 1 cho giá trị lớn nhất, giá trị 2 cho giá trị lớn nhì và tiếp tục cho đến khi kết thúc. Lưu ý rằng nếu có nhiều giá trị bằng nhau, gán giá trị hạng trung bình cho tất cả giá trị đó (ví dụ như việc gán hạng 16.5 cho cả 4 giá trị lớn từ thứ 15 đến 18 trong trường hợp 4 giá trị này bằng nhau)

2. Xác định chênh lệch hạng, $d_i$, của từng cặp quan sát

3. Xác định hệ số tương quan xếp hạng Spearman:

$$
r_S  = 1-\dfrac{6\sum_{i=1}^{n}{d_i^2}}{n(n^2-1)}
$$

Đến đây, chúng ta hoàn toàn có thể sử dụng kiểm định $t$ trong trường hợp cỡ mẫu lớn ($n>30$), hoặc sử dụng một bảng giá trị tới hạn đặc biệt nếu cỡ mẫu nhỏ hơn

# Tests of Independence Using Contingency Table Data

Khi phải làm việc với dữ liệu phân loại hoặc dữ liệu rời rạc, chúng ta không thể sử dụng các phương pháp chúng ta đã làm rõ với nhau để kiểm tra xem liệu các cách phân loại dữ liệu có độc lập với nhau hay không

Giả thuyết không của kiểm định này là: "Hai cách phân loại dữ liệu độc lập với nhau"

Giá trị kiểm định $\chi^2$ được xác định bởi công thức:

$$
\chi^2 = \sum_{i=1}^{I}{\sum_{j=1}^{J}{\dfrac{\left(O_{ij}-E_{ij}\right)^2}{E_{ij}}}}
$$

Trong đó:

- Bậc tự do của phân phối $\chi^2$ là $(I-1)(J-1)$, với $I,J$ lần lượt là số loại dữ liệu ứng với các cách phân loại dữ liệu

- $O_{ij}$ là tần suất kết hợp quan sát được

- $E_{ij}$ là tần suất kết hợp ước tính với giả định $H_0$ được thỏa mãn. Do đó giá trị này được xác định từ ý tưởng xác suất kết hợp. Nếu mẫu có tổng cộng $n$ quan sát, trong đó có $\alpha$ quan sát loại $i$ và $\beta$ quan sát loại $j$, $E_{ij}$ có thể được xác định qua hai hướng tiếp cận

$$
E_{ij} = \dfrac{\alpha \times \beta}{n} = P(i)\times P(j) \times n 
$$

Chỉ tiêu thấp hơn/cao hơn kỳ vọng được xác định thông qua chỉ tiêu "phần dư chuẩn hóa", hoặc đôi khi được gọi là phần dư Pearson:

$$
Standardized \; residual \; = \dfrac{O_{ij}}{E_{ij}}
$$
