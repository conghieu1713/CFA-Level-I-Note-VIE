# Measure of Central Tendency

Xc nh xu hng tp trung (`measure of central tendency`) xc nh ni d liu c tp trung

Php o v tr (`measure of location`) khng ch bao gm cc php o trung tm, m cn quan tm n cc php o minh ha v tr v phn phi ca d liu

## Arithmetic Mean

Trung bnh cng bng tng gi tr ca cc quan st chia cho s quan st
$$
\overline{x} = \dfrac{\sum_{i=1}^{n}{x_i}}{n}
$$

## Trimmed Mean v Winsorized Mean

Trn thc t, cc gi tr ngoi lai c th ch i din cho mt vi gi tr him trong tng th, nhng n cng c th l li trong vic ghi chp d liu hoc gi tr ca quan st ny c to ra t mt tng th khc so vi tng th ca mu.

Gi tr ngoi lai c th khin cc xu hng tp trung b lch. Trong trng hp ny, chng ta cn:

1. Lm sch d liu

2. La chn bin khc vi cng mc ch, hoc s dng cc php chuyn i bin (v d nh ly logarit t nhin)

3. Nu khng th s dng bin thay th hoc chuyn i bin, di y l ba cch x l:

    - Khng lm g c

    - Xa b tt c cc gi tr ngoi lai

    - Thay th cc gi tr ngoi lai bng cc gi tr khc

ng vi cch th hai, chng ta c th s dng trung bnh lc b (`trimmed mean`). i lng ny c tnh bng cch loi b mt t l phn trm gi tr ln nht v gi tr nh nht v tnh trung bnh cng cc gi tr cn li

ng vi cch th ba, chng ta c th s dng trung bnh gn (`winsorized mean`). i lng ny c tnh bng cch thay th mt t l phn trm gi tr nh nht bng mt gi tr c th, v thc hin iu tng t vi t l phn trm gi tr ln nht, sau  tnh trung bnh cng ca mu thu c

## Median

Trung v (`median`) l gi tr nm  chnh gia chui gi tr  c sp xp tng dn hoc gim dn. Trong mt mu gm $n$ quan st, trung v bng gi tr  v tr $\frac{n+1}{2}$ nu $n$ l, v bng trung bnh hai gi tr  v tr $\frac{n}{2}$ v $\frac{n+1}{2}$ nu $n$ chn

## Mode

`Mode` l gi tr c tn s cao nht trong phn phi

Mt phn phi c th c mt `mode`, nhiu `mode` hoc thm ch khng c `mode` no.

- Nu phn phi ch c mt `mode`, n c gi l `unimodal`

- Nu phn phi c hai `mode`, n c gi l `bimodal`

- Nu phn phi c ba `mdoe`, n c gi l `trimodal`

- Nu tt c gi tr trong tp d liu l khc nhau, phn phi khng c `mode` bi khng c gi tr no xut hin nhiu ln hn so vi gi tr khc

Bin lin tc c th khng c `mode`. Tuy nhin, nu d liu c nhm vo cc `interval`, chng ta c th tm thy mt `interval` hoc nhiu `intervals` c tn sut cao nht (`modal interval`)

## Weighted Mean

nh ngha trung bnh trng s (`weighted mean`) c pht trin t phn tch danh mc - chng ta cn quan tm nhiu hn n trng s khc nhau gia cc quan st khc nhau

Vi tp quan st $x_1, x_2, ..., x_n$ v trng s tng ng l $w_1, w_2, ...,w_n$, trung bnh trng s c tnh nh sau:
$$
\overline{x_w} = \sum_{i=1}^{n}{w_i x_i}
$$
Trong , $\sum_{i=1}^n{w_i} = 1$

## Geometric Mean

Trung bnh nhn (`geometric mean`) thng c s dng  tnh t l thay i trung bnh qua thi gian hoc  tnh tc  tng trng ca mt bin. Trong u t, chng ta thng s dng trung bnh nhn  tnh t sut sinh li trung bnh ca mt ti sn/mt danh mc, hoc tnh t l tng trng ca mt bin ti chnh nh li nhun hoc doanh thu

Cng thc tnh trung bnh nhn ca tp gi tr $x_1, x_2, ..., x_n$, trong  $x_i \ge 0 vi $i=1,2,...,n$ l
$$
\overline{x_G}=\sqrt[n]{\prod_{i=1}^{n}{x_i}}
$$

## Harmonic Mean

Trung bnh iu ha (`harmonic mean`) thng c s dng trong u t  tnh chi ph trung bnh
$$
\overline{x_H} = \dfrac{n}{\sum_{i=1}^{n}{\dfrac{1}{x_i}}}
$$

## Q-A-G-H Inequality

Bt ng thc Q-A-G-H th hin mi quan h gia trung bnh cng, trung bnh nhn, trung bnh ton phng (`quadratic average`) v trung bnh iu ha: $\overline{x_Q} \ge \overline{x} \ge \overline{x_G} \ge \overline{x_H}$
$$
\sqrt{\sum_{i=1}^{n}{x_i^2}} \ge \dfrac{\sum_{i=1}^{n}{x_i}}{n} \ge \sqrt[n]{\prod_{i=1}^{n}{x_i}} \ge \dfrac{n}{\sum_{i=1}^{n}{\dfrac{1}{x_i}}}
$$
***Note.*** *Cng thc trung bnh ton phng*
$$
\overline{x_Q} = \sqrt{\sum_{i=1}^{n}{x_i^2}}
$$

# Quantiles

Nhng ngi lm thng k s dng thut ng phn v (`quantile` hay `fractile`)  ch cc im ct phn phi thnh cc khong lin tc vi xc sut bng nhau

## Quartiles, Quintiles, Deciles, Percentiles

Chng ta bit rng trung v chiai phn phi thnh hai na c xc sut nh nhau. Chng ta c th nh ngha cc phn v m chng chia phn phi thnh cc phn nh hn. C th

* T phn v (`Quartiles`): gm 3 im, chia phn phi thnh 4 phn c xc sut bng nhau

* tri gia (`Interquartile` - $IQR$) c tnh bng chnh lch gia t phn v th ba v t phn v th nht: $IQR = Q_3 - Q_1$*

* Ng phn v (`Quintiles`): gm 4 im, chia phn phi thnh 5 phn c xc sut bng nhau

* Thp phn v (`Deciles`): gm 9 im, chia phn phi thnh 10 phn c xc sut bng nhau

* Bch phn v (`Percentiles`): gm 99 im, chia phn phi thnh 100 phn c xc sut bng nhau

Khi x l d liu thc t, chng ta thng phi c lng gi tr ca mt bch phn v no . Gi s chng ta cn c lng gi tr ca bch phn v th $y$ ($P_y) ca mt phn phi m mu ca n c $n$ quan st. Quy trnh xc nh bch phn v ny gm 3 bc:

1. Sp xp li d liu thnh chui theo th t tng dn

2. Xc nh v tr tng i ca phn v ($L_y$) bng cng thc

$$
L_y = (n+1)\dfrac{y}{100}
$$

3. Xc nh gi tr $P_y$.  y c hai trng hp
    
    * Nu $L_y$ l s nguyn, $P_y$ chnh l gi tr th $L_y$ trong chui

    * Nu $L_y$ khng l s nguyn, $P_y$ c xc nh thng qua hai gi tr th $[L_y]$ v $([L_y]+1)$ trong chui bng phng php ni suy tuyn tnh

$$
P_y = X_{[L_y]}\times \left(1 - \left(L_y - [L_y]\right)\right) + X_{[L_y]+1} \times (L_y - [L_y]) 
$$

# Measures of Dispersion

## The Range

Khong gi tr (`Range`) l khong cch gia gi tr ln nht v gi tr nh nht ca d liu

$$
Range = Max - Min
$$

## The Mean Absolute Deviation

 lch trung bnh tuyt i ($MAD$) bng trung bnh gi tr tuyt i  lch ca tt c cc quan st vi gi tr trung bnh ca d liu

$$
MAD = \dfrac{\sum_{i=1}^{n}{\left|X_i - \overline{X}\right|}}{n}
$$

$MAD$ l mt cch gii quyt vn  v du ca cc  lch (r rng chng ta c $\sum_{i=1}^{n}{(X_i - \overline{X})}=0$)

## Sample Variance and Sample Deviation

Mt cch tip cn khc  gii quyt vn  v du ca cc  lch chnh l s dng bnh phng cc  lch. Vi cch ny, chng ta c cc i lng l **phng sai mu** v ** lch chun mu**

### Phng sai mu

$$
s^2 = \dfrac{\sum_{i=1}^{n}{\left(X_i - \overline{X}\right)^2}}{n-1}
$$

###  lch chun mu

$$
s = \sqrt{\dfrac{\sum_{i=1}^{n}{\left(X_i - \overline{X}\right)^2}}{n-1}}
$$

 y, phng sai mu $s^2$ v  lch chun mu $s$ l cc *c lng khng chch* ca phng sai tng th $\sigma^2$ v  lch chun tng th $\sigma$

# Downside Deviation and and Coefficient of Variation

## Downside Deviation

Phng sai/ lch chun t sut sinh li ca ti sn thng c xem nh thc o ri ro ca ti sn. Hai i lng ny tnh n c cc gi tr ln hn v nh hn gi tr trung bnh (ngha l xt c ri ro tng gi v ri ro gim gi). Tuy nhin, cc nh u t thng ch quan tm n ri ro gim gi (`downside risk`). iu  c ngha l phn tch vin cn pht trin cc thc o ri ro gim gi

Trn thc t, chng ta thng lo lng v cc gi tr nh hn mt mc $B$ no  hn l gi tr trung bnh $\overline{X}$.  lch chun di (`target downside deviation` hay `target semideviation`) l mt thc o phn tn cho cc quan st  bn di mc mc tiu

$$
s_{2nd} = \sqrt{\sum_{i=1}^{n}{\dfrac{{\left(Min(X_i-B,0)\right)}^2}{n-1}}}
$$

## Coefficient of Variation

Trong mt s trng hp, chng ta kh c th s dng d lch chun  gii thch mc b bin i tng i gia cc b d liu khc nhau, bi v gi tr trung bnh c s khc bit r rt v/hoc c s khc bit v n v o.  l l do chng ta cn s dng cc thc o phn tn tng i (`relative dispersion`) - i lng m t mc  phn tn so vi mt tham chiu,  gii thch trong cc trng hp ny

H s bin thin (`Coefficient of Variation` - $CV$) l mt thc o phn tn tng i tiu biu

$$
CV = \dfrac{s}{\overline{X}}
$$
